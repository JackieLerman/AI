```python
import spacy
import nltk
import random 
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine
import re
import gzip, json
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter 
```

```python
import os
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
def sim(v1,v2):
    return 1.0 - cosine(v1,v2)
```

```python
# import the PCA module from sklearn
from sklearn.decomposition import PCA
```

```python
# converts a list of words into their word vectors
def get_word_vectors(words):
    return [nlp(word).vector for word in words]
```

```python
def plotpca(words):
    '''plot vector of words in 2D plot by using PCA projection'''
    # intialise pca model and tell it to project data down onto 2 dimensions
    pca = PCA(n_components=2)

    # fit the pca model to our 300D data, this will work out which is the best 
    # way to project the data down that will best maintain the rel
    pca.fit(get_word_vectors(words))

    # Used fitted data to project word vectors to 2D
    word_vecs_2d = pca.transform(get_word_vectors(words))
    # create a nice big plot 
    plt.figure(figsize=(10,10))

    # plot the scatter plot of where the words will be
    plt.scatter(word_vecs_2d[:,0], word_vecs_2d[:,1])

    # for each word and coordinate pair: draw the text on the plot
    for word, coord in zip(words, word_vecs_2d):
        x, y = coord
        plt.text(x, y, word, size= 15)

    # show the plot
    plt.show()
```

```python

poems = []

for line in gzip.open("/Users/jackielerman/Desktop/Final_Proj/gutenberg-poetry-v001.ndjson.gz"):
    poems.append(json.loads(line.strip()))

#print(all_lines)
# Stop words list used for words I do not want to include in analysis, too common, not enough meaning 
stop_words = stopwords.words("english")

extra_stop_words = ["I", "And","the","The","O","In", ".",",","?","As","Yet","Thus","Shall","n","th",";",":","'","Of"]

stop_words.extend(extra_stop_words)


#Line to access a single poem from poems, use number to access specific poems from corpus
#Number corresponds to a 'gid' which can be linked to author and title via gutenberg online database

select_poem = [line['s'] for line in poems if re.search(r'\b42439\b',line['gid'], re.I)]

filtered = []


for line in select_poem:
   # print(line)
    #Tokenize poem into words 
    words = word_tokenize(line)
    
    #remove stop words and place new words into filtered list of words 
    for w in words:
        if w not in stop_words:
            filtered.append(w)
              

emotions_found = []
emotwords = []

#open and clean up our emotions database 
with open('/Users/jackielerman/Desktop/emotions.txt','r') as file:
    for line in file:
        filtered_line= line.replace('\n', '').replace(',','').replace("'",'').strip()
        #Split text file into the word and emotion 
        word,emotion = filtered_line.split(':')

       # if a word from out filtered poem shows up in our database of words attatched to an emotion
       # Add it to emotions_found list
        if word in filtered:
            emotions_found.append(emotion)
            emotwords.append(word)
            

# counter is a list of tuples like this:  [('emotionName',# of times),....]
counter = Counter(emotions_found).most_common(6)

counter1 = Counter(emotwords).most_common(20)


common = []
common1 = []
    
for item in counter:
    common.append(item[0])

for item in counter1:
    common1.append(item[0])
    
    
combine = common + common1

shuffled = random.sample(combine, len(combine))

print(shuffled)
```

```python
plotpca(common1)
plotpca(common)
plotpca(shuffled)
```
